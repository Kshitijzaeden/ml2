Variables in python

```python
import pandas as pd
```

```python
df=pd.read_csv("carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
df_dummies = pd.concat([df,dummies],axis='columns')
df_dummies
```

```python
df_dummies.drop(['Car Model','Audi A5'],axis='columns',inplace=True)
df_dummies
```

```python
X=df_dummies.drop('Sell Price($)',axis='columns')
X
```

```python
y=df_dummies['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

PRICE PREDICTION OF CAR WITH MILEAGE 45000 AND AGE 4YRS USING DUMMY VARIABLES


```python
model.predict([[45000,4,0,0]])
```

HOW MUCH ACCURACY DOES OUR MODEL SHOWS

```python
model.score(X,y)
```
